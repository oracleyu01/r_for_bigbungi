```r
#‚ñ† Ïã§ÏäµÏòàÏ†ú1.  ÏßÅÏóÖÏù¥ SALESMAN Ïù¥Í≥† ÏõîÍ∏âÏù¥ 1200 Ïù¥ÏÉÅÏù∏ ÏÇ¨ÏõêÎì§Ïùò Ïù¥Î¶ÑÍ≥º ÏõîÍ∏âÍ≥º ÏßÅÏóÖÏùÑ Ï∂úÎ†•ÌïòÏãúÏò§ !

emp[ (emp$job=='SALESMAN') & ( emp$sal >= 1200 ), c("ename","sal","job") ]
```
